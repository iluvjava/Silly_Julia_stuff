### Basic Plots
* Line Plots
* Line Plots with style and shit
* Scatter plots with style and shit

```julia
plot(rand(10))
plot!(rand(10))
```

```julia

# title, linewidth, and legend names
plot(rand(10), lw=3, label="Line 1")
plot!(rand(10), label="line2", markdershape=:circle, linestyle=:dot, lw=5)
title!("2 Lines")
xlabel!("X Label")
ylabel!("Y Label")
```

```julia
# current plot 
p = Plots.CURRENT_PLOT
N = 400
scatter(randn(N), randn(N), label="scatter 1")
scatter!(randn(N) .+ 0.5, randn(N) .- 0.5, markersize=4, label="scatter 2")
```

### **More Detauiled Usage**

* Changing the color and style
* Different Layout
* Bar Plot, Box plot, Histogram 

```julia
TheSameDataX = rand(10)
TheSameDataY = rand(10)
plot(rand(100, 4), layout=4) 
```

```julia
l = @layout [
    a{0.5w} [grid(3,3)
             b{0.2h}]
]
plot(
    rand(10, 11),
    layout = l, legend = false, seriestype = [:bar :scatter :path],
    title = ["($i)" for j in 1:1, i in 1:11], titleloc = :right, titlefont = font(8)
)
```

```julia
# This is getting a big crank, let's change the size 
# layout: `;` for next row, ` ` space is for next element in this row
l = @layout [
    a{0.5w} [grid(3,3); b{0.2h}]
]
plot(
    rand(10, 11),
    layout = l, legend = false, seriestype = [:bar :scatter :path],
    title = ["($i)" for j in 1:1, i in 1:11], titleloc = :right, titlefont = font(8)
)
plot!(size=(1000, 500))
```

```julia
# This is getting a big crank, let's change the size 
# layout: `;` for next row, ` ` space is for next element in this row
l = @layout [
    a{0.2h}; 
    [grid(3,3); b{0.2h}]
]
plot(
    rand(10, 11),
    layout = l, legend = false, seriestype = [:bar :scatter :path],
    title = ["($i)" for j in 1:1, i in 1:11], titleloc = :right, titlefont = font(8)
)
plot!(size=(500, 1200))
```
