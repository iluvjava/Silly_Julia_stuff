### Basic Plots
* Line Plots
* Line Plots with style and shit
* Scatter plots with style and shit

```julia
plot(rand(10))
plot!(rand(10))
```

```julia

# title, linewidth, and legend names
plot(rand(10), lw=3, label="Line 1")
plot!(rand(10), label="line2", markdershape=:circle, linestyle=:dot, lw=5)
title!("2 Lines")
xlabel!("X Label")
ylabel!("Y Label")
```

```julia
# current plot 
p = Plots.CURRENT_PLOT
N = 400
scatter(randn(N), randn(N), label="scatter 1")
scatter!(randn(N) .+ 0.5, randn(N) .- 0.5, markersize=4, label="scatter 2")
```
